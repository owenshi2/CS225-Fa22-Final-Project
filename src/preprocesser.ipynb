```python
import pandas as pd
routes = pd.read_csv('./entry/routes.csv')
routes
```

```python
routes = routes.filter(items=["airline" ," source airport", " destination apirport"])
routes
```

```python
codes = pd.read_csv('./entry/airport-codes.csv')
codes.head()
```

```python
codes = codes.dropna(subset=['iso_country', 'iata_code', "municipality"])
codes = codes.filter(items=["iso_country", "municipality", "iata_code"])
codes
```

```python
cities = pd.read_csv('./entry/worldcities.csv')
cities = cities.filter(items=["city_ascii" ,"population"])
cities.head()
```

```python
combined = codes.merge(cities, how='inner', left_on='municipality', right_on='city_ascii')
combined = combined.drop(['municipality'], axis = 1)
combined
```

```python
combined = combined.dropna()
combined = combined.drop_duplicates(subset=['city_ascii'])
combined
```

```python
routes.to_csv('flight-paths.csv', index=False) 
combined.to_csv('cities-data.csv', index=False) 
```
